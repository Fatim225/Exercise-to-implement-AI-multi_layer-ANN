### For activation Functions ###

```python

import numpy as np

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    s = sigmoid(x)
    return s * (1 - s)

def tanh(x):
    return np.tanh(x)

def tanh_derivative(x):
    return 1 - np.tanh(x)**2

```

### Initialize parameters ###

```python
def initParams(X, y):
    n_x = X.shape[0]
    n_y = y.shape[0]
    n_h1 = 3
    n_h2 = 4
    n_h3 = 3

    w1 = np.random.randn(n_h1, n_x) - 0.5
    b1 = np.zeros((n_h1, 1))
    w2 = np.random.randn(n_h2, n_h1) - 0.5
    b2 = np.zeros((n_h2, 1))
    w3 = np.random.randn(n_h3, n_h2) - 0.5
    b3 = np.zeros((n_h3, 1))
    w4 = np.random.randn(n_y, n_h3) - 0.5
    b4 = np.zeros((n_y, 1))

    return [w1, b1, w2, b2, w3, b3, w4, b4]
```

### Forward Propagation ###

```python

def forwardPropagation(X, params):
    w1, b1, w2, b2, w3, b3, w4, b4 = params

    Z1 = np.dot(w1, X) + b1
    A1 = sigmoid(Z1)       # g[1]

    Z2 = np.dot(w2, A1) + b2
    A2 = tanh(Z2)          # g[2]

    Z3 = np.dot(w3, A2) + b3
    A3 = tanh(Z3)          # g[3]

    Z4 = np.dot(w4, A3) + b4
    A4 = tanh(Z4)          # g[4] = output

    epsilon = 1e-15  # Small value to prevent log(0)
    # Clip predictions to avoid 0 and 1
    A4 = np.clip(A4, epsilon, 1 - epsilon)

    return [Z1, A1, Z2, A2, Z3, A3, Z4, A4]

```

### Compute Cost ###

```python
def computeCost(y, A4):
    m = y.shape[1]
    cost=-(1/m)*np.sum(y*np.log(A4)+(1-y)*np.log(1-A4))
    return cost
```

### Backward Propagation ###

```python
#With sigmoid
def backPropagation(X, y, cache, params):
    m = y.shape[1]
    w1, b1, w2, b2, w3, b3, w4, b4 = params
    Z1, A1, Z2, A2, Z3, A3, Z4, A4 = cache

    dZ4 = (A4 - y) * tanh_derivative(Z4)
    dw4 = (1/m) * np.dot(dZ4, A3.T)
    db4 = (1/m) * np.sum(dZ4, axis=1, keepdims=True)

    dZ3 = np.dot(w4.T, dZ4) * tanh_derivative(Z3)
    dw3 = (1/m) * np.dot(dZ3, A2.T)
    db3 = (1/m) * np.sum(dZ3, axis=1, keepdims=True)

    dZ2 = np.dot(w3.T, dZ3) * tanh_derivative(Z2)
    dw2 = (1/m) * np.dot(dZ2, A1.T)
    db2 = (1/m) * np.sum(dZ2, axis=1, keepdims=True)

    dZ1 = np.dot(w2.T, dZ2) * sigmoid_derivative(Z1)
    dw1 = (1/m) * np.dot(dZ1, X.T)
    db1 = (1/m) * np.sum(dZ1, axis=1, keepdims=True)

    return [dw1, db1, dw2, db2, dw3, db3, dw4, db4]
```

### Update parameters ###

```python
def updateParams(params, grads, learning_rate):
    w1, b1, w2, b2, w3, b3, w4, b4 = params
    dw1, db1, dw2, db2, dw3, db3, dw4, db4 = grads
    
    w1 = w1 - learning_rate * dw1
    b1 = b1 - learning_rate * db1
    w2 = w2 - learning_rate * dw2
    b2 = b2 - learning_rate * db2
    w3 = w3 - learning_rate * dw3
    b3 = b3 - learning_rate * db3
    w4 = w4 - learning_rate * dw4
    b4 = b4 - learning_rate * db4
    
    return w1, b1, w2, b2, w3, b3, w4, b4
```

```python
def get_predictions(A4):
    return np.where(A4 >= 0.5, 1, 0)
```

```python
def get_accuracy(y_predicted, y):
    #print(y_predicted, y)
    return np.sum(y_predicted == y) / y.size
```

### Training Function ###

```python
def train(X, y, n_h1 = 3,n_h2 = 4,n_h3 = 3, epochs = 1000, learning_rate = 0.01):
    
    params = initParams( X , y  )
    cost_history = []
    n_print = epochs//20

    for i in range(epochs):
        cache = forwardPropagation(X, params)
        A4 = cache[7]
        cost = computeCost( y , A4 )
        grads = backPropagation(X, y, cache, params)
        params = updateParams(params, grads, learning_rate=0.01)
        cost_history.append(cost)

        if (i+1) % n_print == 0:
            predictions = get_predictions(A4)
            accu = get_accuracy(predictions, y)
            
            print("Iteration", i+1, " | Cost: ", cost, " | Accuracy: ", accu,"\n")
            
   

    w1 = params[0]
    b1 = params[1]
    w2 = params[2]
    b2 = params[3]
    w3 = params[0]
    b3 = params[1]
    w4 = params[2]
    b4 = params[3]
    
    return w1, b1, w2, b2, w3, b3, w4, b4, cost_history
```

```python
w1, b1, w2, b2, w3, b3, w4, b4, cost_history = train(X_norm, y, n_h1 = 3,n_h2 = 4,n_h3 = 3, epochs = 40000, learning_rate = 0.1)
```
