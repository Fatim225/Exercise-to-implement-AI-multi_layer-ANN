```python
#Activation functions
import numpy as np

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    s = sigmoid(x)
    return s * (1 - s)

def tanh(x):
    return np.tanh(x)

def tanh_derivative(x):
    return 1 - np.tanh(x)**2
```

### Initialize Paramters ###

```python
def initParams(X, y, hidden_units): # X and y are already tansposed
    n = X.shape[0]
    n_y = y.shape[0]
    w1 = np.random.rand( hidden_units , n ) - 0.5
    b1 = np.random.rand( hidden_units , 1 ) - 0.5
    w2 = np.random.rand( n_y , hidden_units ) - 0.5
    b2 = np.random.rand( n_y , 1 ) - 0.5
    return w1, b1, w2, b2 
```

### Forward Propagation ###

```python
def forwardPropagation(X, params):
    w1 = params[0]
    b1 = params[1]
    w2 = params[2]
    b2 = params[3]
    
    Z1 = w1.dot(X)+b1
    # In your forwardPropagation, change:
    A1 = np.maximum(0, Z1) 
    Z2 = w2.dot(A1)+b2
    A2 = sigmoid(Z2)

    epsilon = 1e-15  # Small value to prevent log(0)
    # Clip predictions to avoid 0 and 1
    A2 = np.clip(A2, epsilon, 1 - epsilon)

    return Z1, A1, Z2, A2
    
```

### Compute cost ### 

```python
def computeCost(y, A2):
    m = y.shape[1]
    cost = -(1/m) * np.sum(y*np.log(A2)+(1-y)*np.log(1-A2))
    return cost
```

```python
computeCost( y , A2 )
```

### Compute Backward Propagation ###

```python
def backPropagation(X, y, cache, params):

    m = y.shape[1]
    
    w1 = params[0]
    b1 = params[1]
    w2 = params[2]
    b2 = params[3]
    
    Z1 = cache[0]
    A1 = cache[1]
    Z2 = cache[2]
    A2 = cache[3]

    dZ2 = A2-y
    dw2 = (1/m)*(dZ2.dot(A1.T))
    db2 = (1/m)*np.sum(dZ2)
    dZ1 = (w2.T).dot(dZ2)*(1-A2**2)
    dw1 = (1/m)*(dZ1).dot(X.T)
    db1 = (1/m)*(np.sum(dZ1))

    return dw1, db1, dw2, db2
```

### Update the parameters ###

```python
def updateParams(params, grads, learning_rate):

    w1 = params[0]
    b1 = params[1]
    w2 = params[2]
    b2 = params[3]
    
    dw1 = grads[0]
    db1 = grads[1]
    dw2 = grads[2]
    db2 = grads[3]

    w1 = w1- learning_rate * dw1
    b1 = b1- learning_rate * db1
    w2 = w2- learning_rate * dw2
    b2 = b2- learning_rate * db2

    return w1, b1, w2, b2
    
```

```python
params = updateParams(params, grads, learning_rate=0.01)
```

```python
def get_predictions(A2):
    return np.where(A2 >= 0.5, 1, 0)
```

```python
def get_accuracy(y_predicted, y):
    #print(y_predicted, y)
    return np.sum(y_predicted == y) / y.size
```

```python
def train(X, y, hidden_units = 40, epochs = 1000, learning_rate = 0.01):
    
    params = initParams( X , y , hidden_units= 4 )
    cost_history = []
    n_print = epochs//20

    for i in range(epochs):
        cache = forwardPropagation(X, params)
        A2 = cache[3]
        cost = computeCost( y , A2 )
        grads = backPropagation(X, y, cache, params)
        params = updateParams(params, grads, learning_rate=0.01)
        cost_history.append(cost)

        if (i+1) % n_print == 0:
            predictions = get_predictions(A2)
            accu = get_accuracy(predictions, y)
            
            print("Iteration", i+1, " | Cost: ", cost, " | Accuracy: ", accu,"\n")
            
   

    w1 = params[0]
    b1 = params[1]
    w2 = params[2]
    b2 = params[3]
    
    return w1, b1, w2, b2, cost_history
```
